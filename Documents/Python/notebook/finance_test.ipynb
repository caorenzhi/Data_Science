```python
import pandas.io.data as web
from pandas.io import data, wb # becomes
from pandas_datareader import data, wb
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_data={}
for ticker in ['AAPL','IBM','MSFT','GOOG', 'FB']:
    all_data[ticker]=web.get_data_yahoo(ticker, '1/1/2010', '1/1/2017')
price = pd.DataFrame({tic: data['Adj Close'] for tic, data in all_data.iteritems()})
volum = pd.DataFrame({tic: data['Volume'] for tic, data in all_data.iteritems()})
```

```python
price.head()
```

```python
returns = price.pct_change()
returns.tail()
```
